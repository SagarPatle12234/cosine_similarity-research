**What is Cosine Similarity**

Cosine similarity is an mathematical function which is used to find an measure for similarity between the 
two datasets with some features vector. The Mathematical formula for the function is:

cos(theta)=(A.B)/(||A||.||B||)

Here in numerator is the dot product of the two vector between which cosine similarity function has to be performed
and in the denominator is the simple multiplication of the individual magnitude of the same vectors. If the value of 
the cos(theta) is unity then both the vectors are same and if the value of cos(theta) is -1 then both the vectors are 
opposite to each other in direction.This formula is derived from the concept of dot product between the vectors where theta is the angle between them. Okay,let's understand cosine similarity with visualising the graphs of vectors and we will assume the vectors to be in 2d plane for the sake of understanding.
We will use python for the visualization. Let's start!


**Visualizing Cosine Similarity**

```python
import matplotlib.pyplot as plt
import numpy as np

# Define two vectors
vector1 = np.array([2, 3])
vector2 = np.array([4, 1])

# Calculate cosine similarity
cosine_similarity = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
angle_rad = np.arccos(cosine_similarity)
angle_deg = np.degrees(angle_rad)

# Visualization
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Plot the vectors
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1')
plt.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector 2')

# Annotate angle
mid_x, mid_y = (vector1[0] + vector2[0]) / 2, (vector1[1] + vector2[1]) / 2
plt.text(mid_x, mid_y, f'{angle_deg:.2f}°', fontsize=12, color='purple')

# Set plot limits
plt.xlim(-1, max(vector1[0], vector2[0]) + 1)
plt.ylim(-1, max(vector1[1], vector2[1]) + 1)

# Add labels and title
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('Cosine Similarity Visualization')
plt.legend()
plt.show()
```

As You can see here the two vectors with two features where x-axis is our feature 1 and y-axis is our feature 2. The angle between them is a sign of how much similar the two vectors are,smaller the angle the greater is the similarity between the two. In this example you can see the two vectors are having an angle of 42.27 which shows they are similar to each other . Let's see other example where the angle between the two vectors is 0 rad.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define two vectors
vector1 = np.array([2, 3])
vector2 = np.array([4, 6])

# Calculate cosine similarity
cosine_similarity = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
cosine_similarity = np.clip(cosine_similarity, -1, 1)  # Clip to valid range [-1, 1]
angle_rad = np.arccos(cosine_similarity)  # Safely calculate the angle
angle_deg = np.degrees(angle_rad)

# Visualization
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Plot the vectors
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1')
plt.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector 2')

# Annotate angle
mid_x, mid_y = (vector1[0] + vector2[0]) / 2, (vector1[1] + vector2[1]) / 2
plt.text(mid_x, mid_y, f'{angle_deg:.2f}°', fontsize=12, color='purple')

# Set plot limits
plt.xlim(-1, max(vector1[0], vector2[0]) + 1)
plt.ylim(-1, max(vector1[1], vector2[1]) + 1)

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Cosine Similarity Visualization')
plt.legend()
plt.show()
```

**Advantages of Cosine Similarity over Ai algorithms**

1.Requires less computational power compared to other Ai algorithms in case of information retrieval , to compare one feature vector to another feature vector,etc.

2. Orientation-Focused
Cosine similarity measures the cosine of the angle between two vectors, focusing on the direction rather than the magnitude. This makes it particularly effective when comparing the similarity of normalized data (e.g., text data represented as word vectors).

3. Magnitude Independence
It is unaffected by the magnitude of the vectors, making it ideal for applications where the scale of the data isn't significant, such as comparing document content irrespective of length.


4. Sparse Data Handling
Works well with sparse data (e.g., text represented as a bag of words or TF-IDF vectors) since many elements in such representations are zero.


5. Easy to Compute
The formula for cosine similarity is computationally simple, involving dot products and norms, which are efficient to calculate even for high-dimensional data.


6. Normalization by Default
As the formula divides by the product of vector magnitudes, it inherently normalizes the vectors, simplifying preprocessing steps.


7. Versatility
It can be applied to various domains, such as:Text similarity (e.g., finding similar documents or sentences).
Recommendation systems (e.g., collaborative filtering).
Image analysis (e.g., feature matching).


8. Interpretable Range
The similarity score ranges between -1 and 1 (or 0 and 1 for non-negative vectors), making it easy to interpret:
A score of 1 indicates maximum similarity.
A score of 0 indicates orthogonality (no similarity).
A negative score (for vectors with negative values) indicates opposite directions.


9. Robust Against Noise
Since it depends on the orientation, minor variations or noise in vector magnitudes have less impact on the similarity score.

**Guidelines for choosing Cosine Similarity**


Type of Problems where you should use Cosine Similarity:-

  1.To get the similarity between the two feature vector

2. To retrieve information based on features
   
3.High dimensional spare data
  Use Case: Data with many features but sparse values, such as:
  
                1.User preferences in recommendation systems.
                
                2.One-hot encoded categorical data.
                
  Reason: It works well with sparse vectors where many elements are zero, avoiding issues with unnecessary magnitude                    comparisons.
  
Type of Problems where you should not use cosine similarity:-

1.When during comparison between the two vectors magnitude matters one should not use cosine similarity.

2.For non-vectorizable data: If data cannot be represented in vector space, cosine similarity is not applicable.

3.Highly correlated variables: Other measures, like Pearson correlation, might be better suited.

**Hybrid approaches with cosine similarity**

**So going further we will discuss one hybrid approach of Cosine Similarity and CNN to detect the targeted face in webcam.**

For this we will use python.

```python
import cv2
import os
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def load_yolo_model(yolo_path):
    net = cv2.dnn.readNet(os.path.join(yolo_path, "yolov3.weights"), os.path.join(yolo_path, "yolov3.cfg"))
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers().flatten()]
    return net, output_layers

def HOGFeatures(image):
    hog = cv2.HOGDescriptor()
    resized_image = cv2.resize(image, (64, 128))
    hog_features = hog.compute(resized_image)
    return hog_features.flatten()

def OrbFeatures(image):
    orb = cv2.ORB_create()
    keypoints, descriptors = orb.detectAndCompute(image, None)
    return descriptors.flatten() if descriptors is not None else np.array([])

def pad_or_truncate(features, length):
    return np.pad(features, (0, max(0, length - len(features))), 'constant')[:length]

def process_frame(frame, net, output_layers, fixed_length=5000):
    height, width = frame.shape[:2]
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outputs = net.forward(output_layers)

    boxes, confidences, class_ids = [], [], []
    for output in outputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                center_x, center_y = int(detection[0] * width), int(detection[1] * height)
                w, h = int(detection[2] * width), int(detection[3] * height)
                x, y = center_x - w // 2, center_y - h // 2
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    return boxes, confidences, class_ids

def extract_features(image, fixed_length=5000):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    hog_features = HOGFeatures(gray_image)
    orb_features = OrbFeatures(gray_image)
    combined_features = np.concatenate((hog_features, orb_features)) if orb_features.size > 0 else hog_features
    return pad_or_truncate(combined_features, fixed_length)

def HOGAndSurfFeatureExtraction(filepath, fixed_length=5000):
    features = []
    yolo_path = "yolo-coco"
    net, output_layers = load_yolo_model(yolo_path)
    
    for filename in os.listdir(filepath):
        image_path = os.path.join(filepath, filename)
        image = cv2.imread(image_path)
        if image is None:
            continue
        boxes, _, _ = process_frame(image, net, output_layers)
        for x, y, w, h in boxes:
            roi = image[y:y + h, x:x + w]
            if roi.size > 0:
                features.append(extract_features(roi, fixed_length))
    return np.array(features)

yolo_path = "yolo-coco"
sample_images_folder = "Images"
feature_database = HOGAndSurfFeatureExtraction(sample_images_folder)

cap = cv2.VideoCapture(0)
net, output_layers = load_yolo_model(yolo_path)
tracker, box = None, None

while True:
    ret, frame = cap.read()
    if not ret:
        break

    boxes, confidences, _ = process_frame(frame, net, output_layers)
    indices = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    if len(indices) > 0:
        for i in indices.flatten():
            x, y, w, h = boxes[i]
            if x < 0 or y < 0 or x + w > frame.shape[1] or y + h > frame.shape[0]:
                continue
            roi = frame[y:y + h, x:x + w]
            if roi.size == 0:
                continue

            current_features = extract_features(roi)
            if feature_database.size > 0:
                similarities = cosine_similarity([current_features], feature_database)
                if np.max(similarities) > 0.5:
                    cv2.putText(frame, "Match Found", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                    tracker = cv2.TrackerCSRT_create()
                    tracker.init(frame, tuple(boxes[i]))
                    break

    if tracker:
        success, box = tracker.update(frame)
        if success:
            x, y, w, h = map(int, box)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow("Webcam", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

In the above code we have used yolo-coco for object detection,yolo-coco file contains all of the pretrained weights and biases. Images file will contain the images of object that  we have to detect.
The code uses two feature descriptors:

HOG (Histogram of Oriented Gradients):
Extracts shape and edge-related features from the grayscale version of the image.
ORB (Oriented FAST and Rotated BRIEF):
Extracts keypoints and descriptors for texture and structure features.

basically our feature vector will contain HOG and ORB features which we will use to compare. Then, we are using cosine similarity function to compare the feature vectors.

So the algorithm is:-

1.First detect the image with the help of yolo-coco 

2.Extract the features in the form of HOG and ORB

3.do the first two steps with all the images in the Images folder.

4.Now,Compare the feature vector of detected object with feature vectors obtained from all the images in the Images folder.

5.Output:match found if similarity score is between some threshold value and else show nothing.
